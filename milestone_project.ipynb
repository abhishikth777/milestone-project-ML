```python
import pandas as pd

may29 = pd.read_csv("29 may attendance.csv")
may30 = pd.read_csv("30 may attendance.csv")

totmay29 = may29.groupby('Email')['Duration (minutes).1'].sum().reset_index()
totmay30 = may30.groupby('Email')['Duration (minutes).1'].sum().reset_index()

def attendance(duration):
    return "present" if duration >= 70 else "absent"

totmay29["May 29"] = totmay29["Duration (minutes).1"].apply(attendance)
totmay30["May 30"] = totmay30["Duration (minutes).1"].apply(attendance)

totmay29 = totmay29[["Email", "May 29"]]
totmay30 = totmay30[["Email", "May 30"]]

totdays = 2


total_att = pd.merge(
    totmay29,
    totmay30,
    on="Email"
)

total_att = total_att.fillna("absent")

total_att.to_csv("tot attendance.csv", index=False)

att_cols = ['May 29','May 30']
df = pd.read_csv("tot attendance.csv")
df['totpresent'] = (df[att_cols] == "present").sum(axis=1)

percent = (df['totpresent']/2)*100
df['percent']  = percent
def cert(percent):
    if percent >= 80:
        return "Certificate"
    else:
        return "No Certificate"

df['Certificate eligibility'] = df['percent'].apply(cert)
df
```
